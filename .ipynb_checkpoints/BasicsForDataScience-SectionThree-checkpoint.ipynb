### NumPy and K-nearest-neighbor classifier

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.data.shape
```

```python
X = iris.data
y = iris.target
```

#### Implementing KNN by scratch

```python
def myTrainTestSplit(X,y,testSize):
    idx = np.random.permutation(np.arange(y.size))
    test_size = int(np.floor(testSize*y.size))
    
    #below Idx vars are arrays of the indices we will pick from X and y via array-slicing via masking
    testIdx = idx[:test_size]
    trainIdx = idx[test_size:]

    Xtrain = X[trainIdx,:]
    ytrain = y[trainIdx]
    Xtest = X[testIdx,:]
    ytest = y[testIdx]
    
    return Xtrain, Xtest, ytrain, ytest
```

```python
Xtrain,Xtest, ytrain, ytest = myTrainTestSplit(X,y,0.2)
```

```python
## now we have to calculate the distance for each Xtrain 
Distance = np.sum( (Xtrain[:,np.newaxis,:] - Xtest[np.newaxis,:,:])**2, axis=2)
```

```python
Distance.shape
```

```python
nidx = np.argsort(Distance,axis=0)
```

```python
idx = nidx[:10,0]
```

```python
from scipy import stats
mode_results = stats.mode(ytrain[idx])
ytest[0]
```

### Pandas Indexing - implicit vs explicit

```python
# Here we have a series with explicit indexes (1,3,5) but also with implicit indexes (0,1,2)
data = pd.Series(['a', 'b', 'c'], index=[1,3,5])
print(data[1], '\n') # explicit index from the index array, use loc() instead of array notation
print(data[1:2], '\n') # implicit index, use iloc() instead of array notation

print(data.loc[1], '\n')
print(data.iloc[1:2], '\n')
```
