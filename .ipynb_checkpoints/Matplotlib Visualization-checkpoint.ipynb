```python
# with axes limits
plt.figure(figsize=(4,2), dpi=150)
x= np.linspace(-10,10,150)
y1 = np.sin(x)
y2 = np.cos(x)
plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize=(3,2))
ax = plt.axes()
plt.xlim(-20,5)
plt.ylim(-0.5, 0.5)
#or
#plt.axis([-5, 20,-1,1])
ax.plot(x, y1, '-.g')
ax.plot(x, y2, color='red', linestyle='dotted')
```

```python
# title and legend
fig = plt.figure()
ax = plt.axes()
x= np.linspace(-10,10,150)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x, y1, '-.g', label='sin')
ax.plot(x, y2, color='red', linestyle='dotted', label='cos')
plt.title('COS and SIN')
plt.legend(loc='lower left')
plt.xlabel('X values')
plt.ylabel('Y values')
```

```python
# with markers
fig = plt.figure()
ax = plt.axes()
x= np.linspace(-10,10,150)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x, y1,'v', color='blue',linestyle='dashed', markersize=6)
ax.plot(x, y2, '^', color='red', linestyle='dotted', markersize=4)
```

### Scatter Plots

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 100*np.random.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

```python
from seaborn import load_dataset
iris = load_dataset('iris')
colors = np.random.rand(150)
sizes = np.random.rand(150)
plt.figure(figsize=(4,5), dpi=120)
plt.scatter(iris.sepal_width, iris.sepal_length, c= colors, s = sizes, cmap='RdGy')
plt.colorbar()
```

### Contour Plots

```python
def someFunc(X, Y):
    return 2*X*np.sin(Y)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)

X,Y = np.meshgrid(x, y)
Z = someFunc(X,Y)

# drop 'f' for no filling
plt.contourf(X, Y, Z, 30, cmap=plt.cm.coolwarm)
plt.colorbar()
```

### Histogram Plots

```python
d1 = np.random.normal(0, 0.7, 1000)
d2 = np.random.normal(5, 2, 1000)
d3 = np.random.normal(-3, 1.5, 1000)

plt.hist(d1,alpha = 0.35, bins = 50, histtype='stepfilled')
plt.hist(d2, alpha = 0.35, bins = 50, histtype='stepfilled')
plt.hist(d3, alpha = 0.35, bins = 50, histtype='stepfilled')
```

```python
plt.figure(figsize=(4,3), dpi=120)
params = dict(histtype='stepfilled', alpha=0.35, bins=50)
plt.hist(iris.sepal_length,**params)
plt.hist(iris.sepal_width,**params)
plt.hist(iris.petal_length,**params)
plt.hist(iris.petal_width, **params)
```

### SubPlots

```python
fig, ax = plt.subplots(2,3,figsize=(8,3), sharex='col', sharey='row')
x = np.linspace(0,5,20)
y = np.cos(x)
for i in range(2):
    for j in range(3):
        ax[i,j].scatter(x,y)

plt.title('Sub Plots')
plt.legend()
```

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=70)

x = 3
y = 5
fig, ax = plt.subplots(x,y,figsize=(8,3), sharey='col', sharex='row')

k= 0
for i in range(x):
    for j in range(y):
        ax[i,j].imshow(faces.images[k], plt.cm.bone)
        ax[i,j].set(xticks=[], yticks=[])
        k+=1
```

### 3D Intro

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection='3d')

theta = np.linspace(-4*np.pi, 4*np.pi, 100)
Z= np.linspace(-2,2,100)
r = Z**2+1
X = r*np.sin(theta)
Y = r*np.cos(theta)

ax.plot(X,Y,Z,label='3D curve')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
```

### 3D Scatter Plots

```python
fig = plt.figure(figsize=(4,3), dpi=120)
ax = fig.gca(projection='3d')

x = np.random.normal(0,2,100)
y = np.random.normal(2,4,100)
z = np.random.normal(4,6,100)

ax.scatter(x,y,z, c='r', alpha=0.5, marker='s')
```

### 3D Surface Plots

```python
from mpl_toolkits.mplot3d import axes3d
X,Y,Z = axes3d.get_test_data(0.01)

plt.style.use('seaborn')
fig = plt.figure(figsize=(8,4), dpi=120)
ax = fig.gca(projection='3d')

ax.plot_surface(X,Y,Z, cmap=plt.cm.RdGy, alpha=1)
```

```python
lin = np.linspace(-2,2,500)
X,Y = np.meshgrid(lin, lin)
Z = 2*np.cos(X) - 4*np.sin(Y)

fig = plt.figure(figsize=(8,4), dpi=90)
ax = fig.gca(projection='3d')

ax.plot_surface(X,Y,Z, cmap=plt.cm.RdGy, alpha=0.5)
```
