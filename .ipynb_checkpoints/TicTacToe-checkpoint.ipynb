```python
from IPython.display import clear_output

keepPlaying = True
gameInProgress = True
board = []
activePlayer = 0

def display_board():
    global board
    print(f'{board[0]} | {board[1]} | {board[2]}')
    print(f'{board[3]} | {board[4]} | {board[5]}')
    print(f'{board[6]} | {board[7]} | {board[8]}')
    
def is_square_open(testSquare):
    global board
    testSquare -= 1
    isSquareOpen = True if board[testSquare] != 'O' and board[testSquare] != 'X' else False
    
def get_player_input():
    global activePlayer
    validInput = False
    selectedSquare = None
    while validInput != True:
        selectedSquare = input(f'Player {activePlayer}, please enter a number (1-9) for the square you select:')

        isDigit = selectedSquare.isdigit()
        if isDigit:
            selectedSquare = int(selectedSquare) - 1
         
        isInRange = selectedSquare in range(0,9)
        print(f'isDigit: {isDigit} / isInRange: {isInRange}')
        if isDigit and isInRange:
            isSquareOpen = is_square_open(int(selectedSquare))
            
        if isDigit != True or isInRange != True:
            print('Please enter only numbers 1-9')
        elif isSquareOpen == False:
            print('Please select a square that hasn\'t already been selected')
        else:
            validInput = True
        
        #clear_output()
        
    return selectedSquare
        
def update_board(chosenSquare):
    global board, activePlayer
    playerSymbol = 'X' if activePlayer == 0 else 'O'
    board[chosenSquare] = playerSymbol
    display_board()
    
def handle_player_turn():
    global activePlayer
    chosenSquare = get_player_input()
    clear_output()
    update_board(chosenSquare)

def is_game_over():
    global board, activePlayer
    playerSymbol = 'X' if activePlayer == 0 else 'O'
    winningSquares = [[0,1,2], [3,4,5], [6,7,8], [0,3,6], [1,4,7], [2,5,8], [0,4,8], [2,4,6]]
    
    for win in winningSquares:
        testResults = list(filter(lambda square: board[square] == playerSymbol, win))
        if len(testResults) == 3:
            return True
    
    openSquares = list(filter(lambda x: str(board[x]).isdigit(), range(0,9)))
    #is it a tie?
    if len(openSquares) == 0:
        return True
        
    return False

def run_game():
    global board, activePlayer
    gameIsWon = False
    gameWinner = None
    activePlayer = 0
    board = list(range(1,10))
    
    while gameIsWon != True:
        
        handle_player_turn()
        gameIsWon = is_game_over()
        
        if gameIsWon != True:
            # bitwise operator to toggle from 0 to 1 and back
            activePlayer ^= 1
                  
    else:
        gameWinnerMessage = f'Congratulations Player {activePlayer}, you have won this game'
        
    gameInProgress = False

def play_again():
    validResponse = False
    playOn = False
    while validResponse != True:
        playOn = input('Would you like to play again? (Y, N)').lower()

        if playOn == 'Y'.lower() or playOn == 'N'.lower():
            validResponse = True
        else:
            clear_output()
            print('Please enter either Y for Yes or N for No')
            
    return True if playOn.lower() == 'Y'.lower() else False

def game_on():
    clear_output()
    global keepPlaying
    while keepPlaying:
        run_game()

        keepPlaying = play_again()
        clear_output()

    else:
        print('Thanks for playing, please come again')


game_on()
```
