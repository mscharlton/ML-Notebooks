```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bokeh.plotting import figure,output_file,show
```

```python
output_file('bokeh_demo.html')
p = figure(plot_width=800, plot_height=800, title='Bokeh Demo')
p.line(np.random.randn(5),np.random.rand(5), line_width=5)
show(p)
```

```python
from bokeh.io import output_notebook
output_notebook()
p = figure(plot_width=800, plot_height=800, title='Bokeh Demo in Jupyter')
p.line(np.random.randn(5),np.random.rand(5), line_width=5)
show(p)
```

### Multiplot Markers

```python
p = figure(plot_width=400, plot_height=400, title='Bokeh Example', x_axis_label='X axis', y_axis_label='Y axis')
X = np.linspace(-2,2,30)
Y1 = np.sin(X)
Y2 = np.cos(X)
Y3 = Y1 - Y2 + 0.5

p.line(X,Y1,legend_label='sin(X)')
p.circle(X,Y2, legend_label='cos(X)', fill_color='blue', size=3)
p.triangle(X, Y3, legend_label='Y3 values', color='red', size=3)

show(p)
```

### Multiplots Grid Plot

```python
from bokeh.layouts import gridplot
```

```python
X = np.linspace(-2,2,30)
Y1 = np.sin(X)
Y2 = np.cos(X)
Y3 = Y1 - Y2 + 0.5

s1 = figure(plot_width=200, plot_height=200)
s1.line(X,Y1,legend_label='sin(X)', color='firebrick', alpha=0.75)

s2 = figure(plot_width=300, plot_height=300)
s2.circle(X,Y2,legend_label='cos(X)', fill_color='olive', alpha=0.75)

s3 = figure(plot_width=400, plot_height=400)
s3.square(X, Y3, legend_label='Y3 values', color='blue', alpha=0.75)

p = gridplot([[s1,s2],[s3]])
show(p)
```
