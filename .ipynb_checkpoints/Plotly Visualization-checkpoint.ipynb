```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
import plotly.graph_objs as go
```

```python
Z = np.linspace(0,10,50)
X = np.sin(Z)
Y = np.cos(Z)

trace = go.Scatter3d(x=X, y=Y, z=Z,
                                mode='markers',
                                marker=dict(size=12,color=Z, colorscale='RdGy'))

layout = go.Layout(title='Plotly Example')
fig = go.Figure(data=[trace],layout=layout)
fig.show()
```

### Scatter and Surface Plots

```python
X,Y = np.meshgrid(np.linspace(-2,2,50), np.linspace(-2,2,50))
Z = np.cos(X**2 + Y**2)

trace = go.Surface(x=X, y=Y, z=Z, colorscale='viridis')
go.Figure(data = [trace], layout = go.Layout(title='Surface Example')).show()
```
