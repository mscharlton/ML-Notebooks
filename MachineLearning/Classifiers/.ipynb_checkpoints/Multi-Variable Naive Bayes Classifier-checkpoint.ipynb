```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.model_selection import train_test_split
```

## Naive Bayes Classifier  
Uses the assumption of conditional independence of the variables  

Given variables $x_1, x_2, y$,  build a probability model of $P{y|x_1,x_2}$  

$P_{y|x_1, x_2}(y|x_1, x_2) = \frac{f_{x_1,x_2}(x_1, x_2 | y) * P_y(y)}{f_{x_1,x_2}(x_1, x_2)}$  

we'll also make use of the following subsitution:  

$f_{x_1, x_2|y}(x_1, x_2|y) = f_{x_1,y}(x_1|y) * f_{x_2|y}(x_2|y)$  

using the Bayes Rule to break out the individual density equations for each variable  

Using the iris data, we will have 4 random variables to predict the species:
1. petal_length  
2. petal_width  
3. sepal_width  
4. sepal_length

As shown above, we need to calculate $f_{x_1|y}(x_1|y)$ for each of the 4 random variables.  The product of all four will produce the classification probability

```python
# load and setup data values
iris = sns.load_dataset('iris')

x_labels = [ x for x in iris.columns if x != 'species' ] 
y = iris['species'].copy()  

# quantify the species
y[y=='setosa'] = 0
y[y=='versicolor'] = 1
y[y=='virginica'] = 2 
y_size = y.unique().size

iris['species'] = y

# find overall P_y(y) values
y_probs = [ ((y==1).sum() / y.size) for i in range(y_size)]

sns.pairplot(iris, hue='species')
```

```python
def calc_x_prob(x: int, mu: float, sigma: float) -> float:
    return np.exp((-(x-mu)**2)/(2*(sigma**2))) / ((2*np.pi*sigma**2)**0.5)

def prob_y_given_x(x: float, mu: float, sigma: float, y_prob: float) -> float:
    return y_prob * ( calc_x_prob(x, mu, sigma) )

def find_best_y(mu: list, sigma: list, x: float) -> int:
    y_probs_given_x = [ prob_y_given_x(x, mu[i], sigma[i], y_probs[i]) for i in range(y_size) ]
    return y_probs_given_x.index(max(y_probs_given_x))
```

```python
# let's see how testing data looks against the training data
for x_label in x_labels:
    X = iris[x_label]
    x_train,x_test, y_train, y_test = train_test_split(X,y,train_size=0.80)

    mu_values = [ x_train[y_train==i].mean() for i in range(y_size) ]
    sigma_values = [ x_train[y_train==i].std() for i in range(y_size) ]

    # figure out hit ratio
    misses = 0
    test_size = y_test.count()
    for i in y_test.index:
        cat_value = y_test[i] 
        x_test_value = x_test[i]
        pred_value = find_best_y(mu = mu_values, sigma = sigma_values, x = x_test_value) 
        if cat_value is not pred_value:
            misses += 1

    for i in range(y_size):
        # our training
        x_train_data = x_train[y_train==i]
        fx_train = calc_x_prob(x = x_train_data, mu = mu_values[i], sigma = sigma_values[i])
        plt.scatter(x_train[y_train==i], fx_train, label='train cat: '+str(i), color='blue', marker='o', s=5)

        # our predictions
        x_test_data = x_test[y_test==1]
        fx_test = calc_x_prob(x = x_test_data, mu = mu_values[i], sigma = sigma_values[i])
        plt.scatter(x_test_data, fx_test, label='test cat: ' + str(i), color='red', marker='s', s=2)  

    plt_label = f'{x_label} : Total Test: {test_size} / Misses: {misses}' 
    plt.title(plt_label)

    plt.legend()
    plt.show()
```

```python
def predict_y_cat(x: dict) -> int:

    probs_y_given_x = np.zeros((3,4))
    # loop over each possible category value 
    for y_cat_idx in range(y_size):        
            
        # calc the fx(x) probability for each column
        for idx, x_label in enumerate(x_labels):

            X = iris[x_label]
            x_train,x_test, y_train, y_test = train_test_split(X,y,train_size=0.80)
            
            mu = x_train[y_train == y_cat_idx].mean()
            sigma = x_train[y_train == y_cat_idx].std() 
            
            probs_y_given_x[y_cat_idx][idx] = calc_x_prob(x[x_label], mu, sigma)

    # now lets determine which y category has the highest probability
    # taking a short-cut as I know that the P(y) is 0.33
    final_probs = [ x[0]*x[1]*x[2]*x[3] for x in probs_y_given_x]
    pred_category = final_probs.index(max(final_probs))
    return pred_category

total = len(iris)
hits = 0 
misses = 0
for idx,row in iris.iterrows():
    row_dict = row.to_dict()
    actual_value = row.species
    pred_value = predict_y_cat(row_dict)
    if actual_value != pred_value:
        misses += 1
    else:
        hits += 1
accuracy = hits / total
print(f'total trials: {total}')
print(f'hits: {hits}')
print(f'misses: {misses}')
print(f'accuracy: {accuracy}')
```
