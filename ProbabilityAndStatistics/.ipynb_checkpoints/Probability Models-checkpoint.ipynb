# Probability Models

## Probability Law
Represents the probability model describing the known probability of various events, each with its own set of outcomes, excluding the empty set.

$\Omega$ = {1,2,3,4}  <= **sample space** of all possible outcomes  
A = {2,4}  
B = {1,3}  

P(A) >= 0  
P(B) >= 0  

## Probability Axioms or Rules

### General Axiom
- P(A) >= 0  
- P(A $\cup$ B) = P(A) + P(B) given that A and B are disjoint events, ie - A $\cap$ B = { }  ($\phi$)
- P($\Omega$) = 1 where $\Omega$ is the set of all possible outcomes  
- 0 <= P(e) <= 1, where each event has a probability between 0 and 1  

Note: P($\Omega$) = P(A) $\cup$ P($A^c$) where $A^c$ is the compliment of A

### Axiom \#1  
$\Omega$ = A $\cup$ $A^c$  
P($\Omega$) = P(A $\cup$ $A^c$)  
P($\Omega$) = P(A) + P($A^c$)  
1 = P(A) + P($A^c$)  
P(A) = 1 - P($A^c$)  

### Axiom \#2  
When A and B events are not disjoint and overlap:

P(A $\cap$ B) = P(A) + P(B) - P(A $\cap$ B)

When A and B are disjoint events:

P(A $\cap$ B) = 0  

### Axiom \#3  
#### DeMorgan's Law  

(A $\cup$ $B)^c$ = $A^c$ $\cap$ $B^c$  
(A $\cap$ $B)^c$ = $A^c$ $\cup$ $B^c$  

### Continuous Probability Models  
With continuous probability models, P($\Omega$) = 1 no longer holds true and is not valid  

### Conditioning
When a given event is dependent, or conditional, of another event  

P(A|B) is probability of event A given the event B outcome  

P(A|B) = P(A $\cap$ B) / P(B)  

where P(B) > 0  

### Total Probability Theorem  

Given $A_1$, $A_2$, $A_3$, ..... $A_n$ are disjoint events and represent partitions of the sample space ($\Omega$)  
and assuming event B $\subset$ $\Omega$:

P(B) = P(B $\cup$ $A_1$) + P(B $\cup$ $A_2$) + P(B $\cup$ $A_3$) + ..... + P(B $\cup$ $A_n$)  

### Independence  

When P(A|B) = P(A), A and B are independent  
When P(A|B) != P(A), A and B are dependent  

When A and B are independent: 

P(A $\cap$ B) = P(A|B) * P(B) 

#### Conditional Independence  

It can be dependent without conditions, ie - P(A $\cap$ B) != P(A) * P(B)  
But with conditionals, it could become independent, ie = P(A $\cap$ B | C) = P(A|C) * P(B|C)  

#### Bayes Rule  

P(A|B) = ( P(B|A) * P(A) ) / P(B)
