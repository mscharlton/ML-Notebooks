```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#3a
a = np.array([ [4,5,1,0] ])
b = np.array([ [-4, -3, 3, 10] ])

print(a+b)
```

```python
#3b
a = np.array([4,2,0])
b = np.array([6,-4,60])
c = np.array([2,-5,40])

abc = a - b - c
print(abc)

plt.plot([0, a[0]], [0, a[1]], 'r', label='a')
plt.plot([0, b[0]], [0, b[1]], 'g', label='b')
plt.plot([0, abc[0]], [0, abc[1]], 'k', label='a_b_c')
plt.legend()
plt.axis('equal')
plt.axis((-30, 30, -30, 30))
plt.grid()
plt.show()
```

```python
#3c
a = np.array([1,0])
b = np.array([1,2])

a_b = a+b
print(a_b)

plt.plot([0,a_b[0]], [0,a_b[1]])
plt.axis('equal')
plt.plot([-4,4],[0,0], 'k--')
plt.plot([0,0], [-4,4], 'k--')
plt.grid()
plt.axis((-4,4,-4,4))
plt.show()
```

```python
#3d
a = np.array([2,2])
b = np.array([3,4])

a_b = a-b
print(a_b)

plt.plot([0,a_b[0]],[0,a_b[1]], 'g', label='a_b')
plt.axis('equal')
plt.plot([-4,4],[0,0], 'b--')
plt.plot([0,0],[-4,4], 'r--')
plt.grid()
plt.axis((-4,4,-4,4))
plt.show()
```

```python
#3e
a = np.array([[-3,1]])
b = np.array([[3, -1]])

print(a+b)
```

```python
#3f
a = np.array([1,4])
b = np.array([2,8])

a_b = a+b
print(a_b)

c = 20

plt.plot([0,a_b[0]],[0,a_b[1]], 'b', label='a_b')
plt.axis('equal')
plt.plot([-c,c], [0,0], 'k--')
plt.plot([0,0], [-c,c], 'k--')
plt.grid()
plt.axis((-c,c,-c,c))
plt.show()
```

```python
#4a
l = -2
a = np.array([4,3,0])
la = l*a

print(la)

plt.plot([0,a[0]], [0, a[1]], 'r', label='a')
plt.plot([0, la[0]], [0, la[1]], 'g', label='la')
plt.legend()
plt.axis((-9, 9, -9, 9))
plt.grid()
plt.show()
```

```python
#5a

a = np.array([-4,-2])
b = np.array([1,3])

aT = a.T

print(np.dot(aT,b))

#also
print(sum(np.multiply(a, b)))
```

```python
#5b
a = np.array([-4, -2])
aT = a.T #np.transpose
b = np.array([1,3])

print(a), print(aT), print(b)

print(np.dot(aT,b))
```

```python
#5e
a = np.array([0,1,2])
aT = np.transpose(a)

b = np.array([-2,-1, 0])

print(np.dot(aT,b))
```

```python
#6a

a = np.array([1,2])
b = np.array([1,2])
bT = b.T

abT = np.outer(a,bT)

print(abT)
```

```python
#6b
a = np.array([-1,3,0])
b = np.array([0,1,3])
bT = np.transpose(b)
abT = np.outer(a, bT)

print(abT)
```

```python
#6c
a = np.array([1,0,0])
b = np.array([0,1,0])
bT = np.transpose(b)

abT = np.outer(a, bT)

print(abT)
```

```python
#7a
a = np.array([1,3])

mu = 1 / np.linalg.norm(a)

print(mu)

check = np.linalg.norm(a*mu)

print(check)
```

```python
#7b
a = np.array([3,4])
mu = 1 / np.linalg.norm(a)

print(mu)

print(a*mu)

check =  np.linalg.norm(a*mu)

print(check)
```
