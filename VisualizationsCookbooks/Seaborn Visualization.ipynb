```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

df = pd.DataFrame({'X': np.random.normal(0,2,1000),
                              'Y': np.random.normal(4,1,1000),
                              'Z': np.random.normal(-4,10,1000)})
```

```python
plt.figure(figsize=(4,3),dpi=90)
for col in df.columns:
    plt.hist(df[col], density=True, alpha=0.5, bins=30, label=col)
    sns.kdeplot(df[col], shade=True)

plt.legend()
```

```python
for col in df.columns:
    sns.displot(df[col])
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris,hue='species')
```

```python
iris
```

### Relplot - aka Relation Plot, aka Scatter Plot

```python
sns.relplot(x='sepal_width',y='sepal_length',data=iris, hue='species', style='species')
```

### RelPlot Kind Line and Facets

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris, hue='species', style='species', kind='line', ci=None, dashes=True,markers=True)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker', style='smoker',col='time', data=tips)
```

```python
fmri = sns.load_dataset('fmri')
fmri.sort_values('subject')
```

```python
sns.relplot(x='timepoint',y='signal',hue='event', style='event',
                 kind='line',data=fmri[fmri['region']=='parietal'],col='subject',col_wrap=5)
```

### Cat Plots

```python
sns.set(style='ticks',color_codes=True)
#sns.catplot(x='day', y='total_bill', data=tips,jitter=False)
ax = sns.catplot(x='day', y='total_bill', data=tips[tips['smoker']=='No'],kind='box', hue='sex')
ax.fig.suptitle('Tips by Day by Bill by Sex')
```
